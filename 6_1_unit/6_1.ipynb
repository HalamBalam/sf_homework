```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sber_data = pd.read_csv('data/sber_data.csv')
sber_data.head()
```

```python
display(sber_data.tail())
```

```python
sber_data.info()
```

```python
sber_data['price_doc'].max()
```

```python
fig = plt.figure(figsize=(10, 7))
boxplot = sns.boxplot(
    data=sber_data,
    y='ecology',
    x='price_doc',
    orient='h',
    width=0.9
)
boxplot.set_title('Цены по уровням экологии');
boxplot.set_xlabel('Цены');
boxplot.set_ylabel('Уровень экологии');
boxplot.grid()
```

```python
jointplot = sns.jointplot(
    data=sber_data, 
    y='price_doc', 
    x='kremlin_km',
    height=8,
)
```
